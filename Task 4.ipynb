```python
import pandas as pd
import numpy as np
from itertools import repeat
```

```python
features = pd.read_csv("features.txt", sep=' ', names=['id', 'cols'])
subject_train = np.loadtxt("subject_train.txt")
subject_test = np.loadtxt("subject_test.txt")
x_train = np.loadtxt("X_train.txt")
y_train = np.loadtxt("Y_train.txt")
x_test = np.loadtxt("X_test.txt")
y_test = np.loadtxt("Y_test.txt")
```

**1. Merges the training and the test sets to create one data set.**

```python
x_train = np.append(x_train, subject_train.reshape(-1,1), axis=1)
x_train = np.append(x_train, y_train.reshape(-1,1), axis=1)

x_test = np.append(x_test, subject_test.reshape(-1,1), axis=1)
x_test = np.append(x_test, y_test.reshape(-1,1), axis=1)

X = pd.DataFrame(np.append(x_train, x_test, axis=0), columns=features.cols.values.tolist()+['subject', 'activity'])
X
```

**2. Extracts only the measurements on the mean and standard deviation for each measurement.**

```python
X_mean_std = X[ features[features.cols.str.contains('mean') | features.cols.str.contains('std')].cols.values]
X_mean_std.head()
```
